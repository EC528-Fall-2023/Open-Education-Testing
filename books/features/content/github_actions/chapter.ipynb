# Creating an OPE Textbook

Creating an OPE textbook Documentation

## Steps

1. Create an empty repsitory
2. Download [OPEEFFORT tools repository](https://github.com/OPEFFORT/tools) 
```shell

cd tools-main

```
3. Generate an SSH key, using the email associated with your github account

```shell

ssh-keygen -t rsa -b 4096 -C 'your_email@bu.edu'

```
4. Use this command to look at the content of your key, then copy and add the SSH key to your associated github account: Settings > SSH and GPG keys

```shell

cat ~/.ssh/config/id_rsa.pub

```
5. Make sure you are able to use github commands on your linux machine, if not run the commands below

```shell

sudo apt update
sudo apt install git

```
6. Within the tools directory, clone the empty repository you created with the command below: fill in the username and repository name

```shell

git clone git@github.com:your_github_username/your_github_empty_repo.git

```
7. Install the OPE Command Line Tool by running the install.sh file in the tools directory as shown below

```shell

./install.sh
```
Paste the outputted export path line in your command-line

Verify installation by running the ope command

```shell

ope

```
8. Run this command (including your own project repo name and repo url) to create a new project using the empty repository URL as shown below
You can find your repo’s SSH URL when you the green <> Code button on your github repo
This will create subdirectories of books, containers, and content in your github repo

```shell

ope new_project your_project_name your_repo_url
```
9. Create a new book in your repository, replacing your_book_name

```shell

ope new_book your_book_name

```
10. Move to your book's content directory, edit the configure file to change book settings (title, author, logo, URL to publish to, and jupyter extensions enabling)

If you want to change your textbook's logo on the top left corner of your book: change the logo .png name and place a new .png in the content directory

```shell

nano your_book_name

```
If your are looking for which line and column an error is on, you can use the command below to be able to see the line and column numbers

```shell

nano -c your_book_name

```
11. In order to see all the changes you are making in these files you must publish the book using the command below, there is a few reasons this command may not run yet 

```shell

make pub

```
To do this, make sure you are in your_book_name directory, you should be able to see your Makefile in this directory (you can verify the pub target exists at the top of the Makefile, when you open it)

If you are receiving an error message regarding ghp-import, install the ghp-import command which is used to publish content to GitHub Pages

```shell

pip install ghp-import

```
To avoid more errors and make ghp-import globally accessible – make sure to add ‘ghp-import’s binary path to your system’s path by modifying your shell (.bashrc or .zshrc) as shown below

```shell

export PATH= '/usr/local/bin:$PATH'

```
Reload your terminal to apply changes

```shell

source ~/.bashrc
```
Verify command installation by running the command below

```shell

ghp-import --version
If the make pub command is still not running, run the command below to 

```shell

source ~/.bashrc
```
